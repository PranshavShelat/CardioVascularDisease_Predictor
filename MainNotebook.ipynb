```python
from sklearn.model_selection import train_test_split
x=df.drop('cardio',axis=1)
y=df['cardio']
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("Training set size:", x_train.shape)
print("Test set size:", x_test.shape)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc_model = SVC()
svc_model.fit(x_train, y_train)
y_pred_svc=svc_model.predict(x_test)
svc_accuracy=accuracy_score(y_test,y_pred_svc)
print("SVC Accuracy: ",round(svc_accuracy,4))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train, y_train)
y_pred_knn=knn_model.predict(x_test)

knn_accuracy=accuracy_score(y_test, y_pred_knn)
print("KNN Accuracy: ", round(knn_accuracy,4))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr_model=LogisticRegression(max_iter=10000)

lr_model.fit(x_train,y_train)
y_pred_lr=lr_model.predict(x_test)
lr_accuracy=accuracy_score(y_test, y_pred_lr)
print("Logistic Regression Accuracy: ",round(lr_accuracy,4))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dt_model=DecisionTreeClassifier()
dt_model.fit(x_train,y_train)

y_pred_dt=dt_model.predict(x_test)
dt_accuracy=accuracy_score(y_test,y_pred_dt)
print("Decision tree accuracy: ", round(dt_accuracy,4))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf_model=RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
y_pred_rf=rf_model.predict(x_test)

rf_accuracy=accuracy_score(y_test,y_pred_rf)
print("Random forest accuracy: ", round(rf_accuracy,4))
```

```python
finalModel=LogisticRegression(max_iter=10000)
finalModel.fit(x_train,y_train)
y_pred_final=finalModel.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

final_accuracy=accuracy_score(y_test,y_pred_final)
print("Final model accuracy: ",round(final_accuracy,4))

print("\nConfusion matrix:",confusion_matrix(y_test,y_pred_final))

report=classification_report(y_test,y_pred_final,output_dict=True)
report_df=pd.DataFrame(report).transpose()
print("\nClassification Report:\n")
display(report_df.round(2))

print("\nConfusion matrix: \n")
cm=confusion_matrix(y_test,y_pred_final)
cm_df=pd.DataFrame(cm).transpose()
cm_df.columns=['Predicted 0','Predicted 1']
cm_df.index=['Actual 0','Actual 1']
display(cm_df.round(2))

```
